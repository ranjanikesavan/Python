```python
import re
```

# Reading the tweet File

```python
pulitzerWords = open('pulitzer.txt','r').read().split(' ')
```

# Storing Stop words, Positive words and Negative words

```python
#Create empty lists for Stop words, Positive words and Negative words
stopWords = []
posWords = []
negWords = []
```

```python
#Read the file, split the file into individual words, change the case to lower for efficient comparison with the tweet words and then add the words to the corresponding list
stopWords.extend(word.lower() for word in open('stop-words.txt','r').read().split())
posWords.extend(word.lower() for word in open('positive-words.txt','r').read().split())
negWords.extend(word.lower() for word in open('negative-words.txt','r').read().split())
```

# Eliminating Stop words

```python
# Initialize a new list to store the clean words after eliminating stop words
pulitzerWords1 = []

# If a word in the tweet file is present in the stop words, eliminate it. 
for word in pulitzerWords:
    if word.lower() not in stopWords:
            pulitzerWords1.extend(re.findall('^[a-z]+',word.lower())) # Consider only words that have alphabet
```

# Scoring the sentiment

```python
pos = 0
neg = 0
overall_value = 0
for word in pulitzerWords1:
    # If the word is present in the positive word list, increment the postive sentiment score by 1
    if word in posWords:
        pos+=1
        overall_value += 1.0
    # If the word is present in the negative word list, increment the negative sentiment score by 1
    elif word in negWords:
        neg+=1
        overall_value -= 1.0

print "positive score is: %d" % pos
print "negative score is: %d" % neg
print "overall score is : %d" % overall_value
```

# Finding the overall sentiment

```python
# To find the overall sentiment, compare the positive and negative score.
if pos > neg:
    sentiment = "positive"
    if pos/neg > 2: #Check if the positive score is atleast twice more the negative score
        print "The tweets about Pulitzer are strongly positive"
    else:
        print "The tweets about Pulitzer are weakly positive"
elif neg > pos: 
    sentiment = "negative"
    if neg/pos > 2: #Check if the negative score is atleast twice more the positive score
        print "The tweets about Pulitzer are strongly negative"
    else:
        print "The tweets about Pulitzer are weakly negative"
else:
    print "The tweets about Pulitzer are neutral" 

```

# Alternate Approach to finding the overall sentiment

```python
# Find the number of lines in the tweet file
with open("pulitzer.txt") as f:
    linecount = len(f.readlines())
print "Line count is: %d" % linecount

#Use line count as a proxy for number of tweets
#Check if the ratio between overall score and number of tweets is greater than 0.5
if abs(overall_value/linecount) > 0.5:
    print "The tweets about Pulitzer are strongly %s" %sentiment
else:
    print "The tweets about Pulitzer are weakly %s" %sentiment
```
